# Hamiltoniano de Jaynes-Cumming

image.png

```python
wc = 1.0  # frequência da cavidade
wa = 1.0  # frequeência do átomo
g = 0.1  # força do acoplamento

# Operador de criação da cavidade
a = tensor(destroy(5), qeye(2))

# operadores no átomo
sz = tensor(qeye(5), sigmaz())  # sigma z
sm = tensor(qeye(5), destroy(2)) # sigma menos

# O hamiltoniano
H = wc * a.dag() * a - 0.5 * wa * sz + g * (a * sm.dag() + a.dag() * sm)

print(H)
```

#Evolução Unitária

Podemos estudar a evolução de um sistema utilizando a função mesolve, abreviação de Master Equation, ela dá a solução para uma equação mestra. Pode conter dinâmica dissipativa ou recair na equação de Schrodinger ou de von Neuman.

```python
#Seja um Hamiltoniano H = X

H = sigmax()

#Estado inicial

psi0 = basis(2,0)

# Fatiando um intervalo de tempo que a função guardará o vetor de estado
tlist = np.linspace(0, 10, 100)

result = mesolve(H, psi0, tlist, [], [])
```

```python
help(mesolve)
```

```python
print(result)
```

O resultado é uma lista que guarda 100 vetores de estado de cada intervalo de tempo da evolução.

```python
len(result.states)
```

```python
# Pegando o último vetor de estado da lista
print(result.states[-1])
```

```python
result.states[-1].norm()
```

# Valores Esperados

É possível calcular valores esperados de operadores.

```python
# função expect(operador que eu quero o valor esperado, estado )
expect(sigmaz(), result.states[-1])
```

Podemos calcular os valores esperados durante toda evolução:

```python
# uma lista de estados no segundo argumento, e a função me retorna uma lista com todos valores esperadores
expect(sigmaz(), result.states)
```

```python
help(plt.subplots())
```

```python
fig, axes = plt.subplots(1, 1)

axes.plot(tlist, expect(sigmaz(), result.states))

axes.set_xlabel(r"$t$", fontsize=20)
axes.set_ylabel(r"$\left<\sigma_z\right>$", fontsize=20);
```

Podemos passar para função "mesolve" uma lista de operadores que desejamos calcular valores esperados durante a evolução.

```python
result = mesolve(H, psi0, tlist, [], [sigmax(), sigmay(), sigmaz()])
```

```python
fig, axes = plt.subplots(1, 1)

axes.plot(tlist, result.expect[2], label=r"$\left<\sigma_z\right>$")
axes.plot(tlist, result.expect[1], label=r"$\left<\sigma_y\right>$")
axes.plot(tlist, result.expect[0], label=r"$\left<\sigma_x\right>$")

axes.set_xlabel(r"$t$", fontsize=20)
axes.legend(loc=2);
```
